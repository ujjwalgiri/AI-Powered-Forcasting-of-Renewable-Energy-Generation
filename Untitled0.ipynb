```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/updated_dataset_2025.csv"  # Update the path if needed
thesis_dataset = pd.read_csv(file_path)

# Ensure 'total_renewable_generation' exists
if 'total_renewable_generation' not in thesis_dataset.columns:
    raise KeyError("Column 'total_renewable_generation' not found in dataset.")

# Split dataset into train (60%)
train_size = int(0.6 * len(thesis_dataset))
train = thesis_dataset[:train_size]

# Generate dummy predictions (Replace with actual model predictions later)
total_train_pred = train['total_renewable_generation'] * 0.95

# Create DataFrame for visualization
train_df = train[['total_renewable_generation']].copy()
train_df['Predicted'] = total_train_pred
train_df['Predicted Percentage'] = train_df['Predicted'] / train_df['total_renewable_generation']

# Plot trend chart
plt.figure(figsize=(12, 6))

# Actual values (entire dataset)
plt.plot(thesis_dataset.index, thesis_dataset['total_renewable_generation'], label='Actual', color='blue')

# Train predictions (only for training data)
plt.plot(train_df.index, train_df['Predicted'], color='orange', linestyle='dashed', label='Train Predicted')

# Labels and title
plt.xlabel('Index')
plt.ylabel('Total Renewable Generation (MW)')
plt.title("Actual vs Train Predictions - Renewable Energy Trend")
plt.legend(loc="upper right")
plt.grid(True)

# Show the plot
plt.show()

# Display first few rows with predicted percentage
print(train_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/updated_dataset_2025.csv"
df = pd.read_csv(file_path)

# Convert timestamp column to datetime
df["utc_timestamp"] = pd.to_datetime(df["utc_timestamp"], errors='coerce')

# Check for invalid dates
invalid_dates = df[df["utc_timestamp"].isna()]
if not invalid_dates.empty:
    print("Invalid timestamps found:")
    print(invalid_dates)
    # Optionally drop rows with invalid dates
    df = df.dropna(subset=["utc_timestamp"])

# Extract year-month for grouping
df["year_month"] = df["utc_timestamp"].dt.to_period("M").astype(str)

# Aggregate total renewable generation by month
monthly_data = df.groupby("year_month")["total_renewable_generation"].sum().reset_index()

# Create a simple moving average as a predicted value (window of 3 months)
monthly_data["predicted_generation"] = monthly_data["total_renewable_generation"].rolling(window=3, min_periods=1).mean()

# Ask for user input to select a specific month
selected_month = input("Enter the month you want to analyze (YYYY-MM): ")

# Filter data for the selected month
filtered_data = monthly_data[monthly_data["year_month"] == selected_month]

if not filtered_data.empty:
    # Plot the graph for the selected month
    plt.figure(figsize=(8, 5))
    plt.bar(["Actual Generation", "Predicted Generation"],
            [filtered_data["total_renewable_generation"].values[0],
             filtered_data["predicted_generation"].values[0]],
            color=['blue', 'orange'])

    plt.xlabel("Type")
    plt.ylabel("Energy Generation")
    plt.title(f"Actual vs. Predicted Renewable Energy Generation ({selected_month})")
    plt.grid(axis="y")
    plt.show()
else:
    print(f"No data found for {selected_month}. Please try another month.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/updated_dataset_2025.csv"  # Replace with your dataset's path
df = pd.read_csv(file_path)

# Convert timestamp column to datetime
df["utc_timestamp"] = pd.to_datetime(df["utc_timestamp"], errors='coerce')

# Drop invalid rows with NaT in utc_timestamp
df = df.dropna(subset=["utc_timestamp"])

# Extract year-month for grouping
df["year_month"] = df["utc_timestamp"].dt.to_period("M").astype(str)

# Aggregate total renewable generation by month
monthly_data = df.groupby("year_month")["total_renewable_generation"].sum().reset_index()

# Create a simple moving average as a predicted value (window of 3 months)
monthly_data["predicted_generation"] = monthly_data["total_renewable_generation"].rolling(window=3, min_periods=1).mean()

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(monthly_data["year_month"], monthly_data["total_renewable_generation"], marker='o', label="Actual Generation", color='blue')
plt.plot(monthly_data["year_month"], monthly_data["predicted_generation"], marker='o', linestyle='--', label="Predicted Generation", color='orange')

# Customize the graph
plt.xlabel("Month")
plt.ylabel("Energy Generation")
plt.title("Actual vs. Predicted Renewable Energy Generation (Monthly)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```
